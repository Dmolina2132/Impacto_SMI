```python
def salaries_process(df, separate_months = False):
    #Rename columns
    df = df.rename(columns=salary_rename_dict)

    #Remove the first to numbers of the ccaa and trim
    # df['ccaa'] = df['ccaa'].apply(lambda x: x[2:] if (x != "Nacional") and (x != "Total Nacional") else x).str.strip()


    #Numbers are in string format, we have to change that
    if df['Total'].dtype == 'object':
        df.loc[(df['Total']=="..") | (df['Total']==".") , 'absence_data'] = 1
        df.loc[(df['Total']==".." ) | (df['Total']=="."), 'Total'] = 0
        df['Total'] = df['Total'].str.replace(".", "").str.replace(",", ".").astype(float)


    return df
```

```python
folder_ipc = "../data/ipc/"
folder_ipri = "../data/ipri/"

gasto_hogar_pdf = pd.read_csv(folder_ipc + "gasto_por_hogar.csv", sep = ";", decimal=",", thousands=".")
ipc_df = pd.read_csv(folder_ipc + "ipc.csv", sep = ";", decimal=",", thousands=".")
ipri_df = pd.read_csv(folder_ipri + "ipri_ccaa_destino_economico.csv", sep = ";", decimal=",", thousands=".")
```

```python
ipc_dict = {'Tipo de jornada': 'jornada', 'Comunidades y Ciudades Autónomas': 'ccaa', 'Comunidad autónoma de residencia': 'ccaa', 'Periodo': 'periodo',
                         'Sectores de actividad': 'sector', 'Gastos totales, medios y distribución porcentual': "tipo_gasto",
                         "Grupos de gasto (2 dígitos)": "grupo_gasto", "Grupos ECOICOP":"grupo_indice", "Tipo de dato": "tipo_dato",
                         "Índice y tasas": "tipo_dato", "Destino económico de los bienes":"destino_economico"}
```

```python
def ipc_ipri_process(df, separate_months = False):
    #Rename columns
    df = df.rename(columns=ipc_dict)

    #Remove the first to numbers of the ccaa and trim
    df['ccaa'] = df['ccaa'].apply(lambda x: x[2:] if (x != "Nacional") and (x != "Total Nacional") else x).str.strip()

    if separate_months:
        df['mes'] = df['periodo'].str[-2:].astype(int)
        df['año'] = df['periodo'].str[:4].astype(int)
        df['periodo_fecha'] = pd.to_datetime(df["periodo"], format="%YM%m")

    #Numbers are in string format, we have to change that
    if df['Total'].dtype == 'object':
        df.loc[(df['Total']=="..") | (df['Total']==".") , 'absence_data'] = 1
        df.loc[(df['Total']==".." ) | (df['Total']=="."), 'Total'] = 0
        df['Total'] = df['Total'].str.replace(".", "").str.replace(",", ".").astype(float)


    return df
```

```python

afiliacion_dict = {'Regímenes': 'regimen', 'comunidades autónomas': 'ccaa', 'Seccion cnae': 'seccion general', 'Periodo': 'periodo',
                         'Afiliados medios R. General y RE. Autonomos': 'total'}

def afiliacion_process(df):
    #Rename columns
    df = df.rename(columns=afiliacion_dict)

    #Remove the first to numbers of the ccaa and trim
    df['ccaa'] = df['ccaa'].apply(lambda x: x[3:] if x != "Total Nacional" else x).str.strip()

    #Extract the month
    df['mes'] = df['periodo'].astype(str).str[-2:].astype(int)
    df['periodo'] = df['periodo'].astype(str).str[:4].astype(int)

    #Numbers are in string format, we have to change that
    if df['total'].dtype == 'object':
        df.loc[(df['total']=="..") | (df['total']==".") , 'absence_data'] = 1
        df.loc[(df['total']==".." ) | (df['total']=="."), 'total'] = 0
        df['total'] = df['total'].str.replace(".", "").str.replace(",", ".").astype(float)


    return df
```

```python
valid_ccaa = [  'ANDALUCÍA',
                'ARAGÓN',
                'ASTURIAS, PRINCIPADO DE',
                'BALEARS, ILLES',
                'CANARIAS',
                'CANTABRIA',
                'CASTILLA Y LEÓN',
                'CASTILLA - LA MANCHA',
                'CATALUÑA',
                'COMUNITAT VALENCIANA',
                'EXTREMADURA',
                'GALICIA',
                'MADRID, COMUNIDAD DE',
                'MURCIA, REGIÓN DE', 
                'NAVARRA, COMUNIDAD FORAL DE',
                'PAÍS VASCO',
                'RIOJA, LA',
                'CEUTA',
                'MELILLA',
                'TOTAL NACIONAL']

comunidades_mapping = {
   'andalucía': 'Andalucía',
   'aragón': 'Aragón', 
   'asturias, principado de': 'Asturias, Principado de',
   'balears, illes': 'Balears, Illes',
   'canarias': 'Canarias',
   'cantabria': 'Cantabria',
   'castilla y león': 'Castilla y León',
   'castilla - la mancha': 'Castilla - La Mancha', 
   'cataluña': 'Cataluña',
   'comunitat valenciana': 'Comunitat Valenciana',
   'extremadura': 'Extremadura',
   'galicia': 'Galicia',
   'madrid, comunidad de': 'Madrid, Comunidad de',
   'murcia, región de': 'Murcia, Región de',
   'navarra, comunidad foral de': 'Navarra, Comunidad Foral de',
   'país vasco': 'País Vasco',
   'rioja, la': 'Rioja, La',
   'ceuta': 'Ceuta y Melilla',
   'melilla': 'Ceuta y Melilla',
   'total nacional': 'Total Nacional'
}
```

```python
pib_rename = {'Comunidad Autónoma': 'ccaa'}
```

```python
def abs_pib_process(df: pd.DataFrame) -> pd.DataFrame:
    last_year = 0
    #Remove columns with (A) or (P) inside
    to_remove = [column for column in df.columns if isinstance(column,str) and (('(A)' in column) or ('(P)' in column))]
    df.drop(to_remove, axis=1, inplace=True)
    for column in df.columns:
        if df[df[column].notna()].shape[0]==0:
            df.drop(column, axis=1, inplace=True)
        else:
            if isinstance(column,int):
                last_year = column
            if column != "Comunidad Autónoma":
        
                df.rename(columns={column:df[column].values[0] + '-' +str(last_year)}, inplace=True)
    df.drop(0,axis=0, inplace=True)
    #Exclude invalid rows
    df = df[df['Comunidad Autónoma'].isin(valid_ccaa)]
    
    #Rename columns
    df.rename(columns=pib_rename, inplace=True)

    #Convert to lower the ccaa
    df['ccaa'] = df['ccaa'].str.lower().replace(comunidades_mapping)

    pivot_columns = [column for column in df.columns if '-' in column]

    #Pivot year column and use the last value in the name
    pivot_columns = [column for column in df.columns if '-' in column]
    pivot_values = [val.replace('(P)','').replace('(A)', '').strip() for val in pivot_columns]
    dict_column_val = dict(zip(pivot_columns,pivot_values))
    df = pd.melt(df, id_vars=['ccaa'], value_vars=pivot_columns)
    df.rename(columns= {'variable': 'periodo_tipo_dato', 'value': 'valor'}, inplace=True)
    df['periodo_tipo_dato'] = df['periodo_tipo_dato'].replace(dict_column_val)
    df['periodo'] = df['periodo_tipo_dato'].apply(lambda x: int(x.split('-')[1]))
    df['tipo_dato'] = df['periodo_tipo_dato'].apply(lambda x: x.split('-')[0])
    df.drop('periodo_tipo_dato', axis=1, inplace = True)
    return df



def pib_process(df: pd.DataFrame) -> pd.DataFrame:
    #Exclude invalid
    for column in pib_var.columns:
        if "Unnamed" in column:
            pib_var.drop(column, axis=1, inplace=True)
    #Exclude invalid rows
    df = df[df['Comunidad Autónoma'].isin(valid_ccaa)]
    
    #Rename columns
    df.rename(columns=pib_rename, inplace=True)

    #Convert to lower the ccaa
    df['ccaa'] = df['ccaa'].str.lower().replace(comunidades_mapping)

    #Pivot year column and use the last value in the name
    year_columns = [column for column in df.columns if '20' in column]
    year_values = [val.replace('(P)','').replace('(A)', '').strip()[0:5] for val in year_columns]
    dict_column_val = dict(zip(year_columns,year_values))
    df = pd.melt(df, id_vars=['ccaa'], value_vars=year_columns)
    df.rename(columns= {'variable': 'periodo', 'value': 'variacion'}, inplace=True)
    df['periodo'] = df['periodo'].replace(dict_column_val)
    return df
```
